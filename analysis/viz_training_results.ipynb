```python
import json
import matplotlib.pyplot as plt
```

```python
training_results_path = "../results/checkpoints/checkpoints_neural_network_based_safety_metric_sota/logfile.json"
with open(training_results_path) as f:
    data = json.load(f)
print(data.keys())
```

```python
font = {'size': 15}
plt.rc('font', **font)
fig, axes = plt.subplots(1, 2, figsize=(8, 4), dpi=100)

axes[0].plot(data['epochs'], data['train_loss'])
axes[0].plot(data['epochs'], data['val_loss'])
axes[0].set_xlabel('Number of epochs')
axes[0].set_ylabel('Binary cross entropy loss')
axes[0].legend(['Training', 'Validation'])

axes[1].plot(data['epochs'], data['train_acc'])
axes[1].plot(data['epochs'], data['val_acc'])
axes[1].set_xlabel('Number of epochs')
axes[1].set_ylabel('R-squared accuracy')
axes[1].legend(['Training', 'Validation'])

for ax in axes:
    for d in ['top','bottom','left','right']:
        ax.spines[d].set_linewidth(1.5)
    ax.tick_params(width=1.5)
    ax.tick_params(top='on', bottom='on', left='on', right='on')
fig.tight_layout()
plt.savefig("logfile.png")
```
